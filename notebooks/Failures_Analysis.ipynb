```python
import sys
sys.path.append('../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.decomposition import PCA

from parse_logs.parse_base import LogDir
from parse_logs.extract_experiment_result import parse_experiment_result
from parse_logs.parse_coordinator_logs import read_design

from parse_logs.parse_coordinator_logs import get_flat_design

sns.set(rc={"figure.figsize":(16, 8)})
# parser.root_data_path = '../tests/data'
#trial_run_objects = parser.get_trial_runs(
#    exp_run_code='experiment_2021_04_01_16_20_00_run_1')
exec_code='experiment_2021_07_29_15_33_17_run_1'
LogDir.base_data_path = '../data'
trial_run_objects = parse_experiment_result(exec_code='experiment_2021_07_29_15_33_17_run_1')

# load trials run to dict (it can take a while)
# sort and instantiate a dataframe
trial_run_dicts = [trial_run.to_dict() for trial_run in trial_run_objects]
# trial_run_dicts
```

```python
tr_df = pd.DataFrame.from_records(trial_run_dicts)
trial_run_dicts.sort( key=lambda tr: tr['exec_group']+tr['code'])
# trial_run_dicts
def get_treatment(row):
    return 'planned' if row['code'][-1] == 'p' else 'baseline'

tr_df['treatment'] = tr_df.apply(get_treatment,  axis=1)
tr_df

#generate a file with scenario with failures
tr_df[(tr_df.end_state != 'success') & (tr_df.treatment == 'planned')].to_csv('failures.csv', index=False)
```

```python
# load tasks
from evaluate.trial_result import TrialResultEvaluator
tre = TrialResultEvaluator()

task_results = tre.task_results(exec_code=exec_code)

# trs = [tr for tr in task_results]
tar_df = pd.DataFrame.from_records(task_results)
tar_df['treatment'] = tar_df.apply(get_treatment,  axis=1)
tar_df
```

```python
# improve readability
tar_df.drop(columns=['parameters', 'exec_code'])
```

```python
tr_df = tr_df.sort_values(by=['exec_group', 'scenario_id']).reset_index(drop=True)

ftr_df = tr_df[(tr_df.end_state != 'success') & (tr_df.treatment == 'planned')]
ftr_df = ftr_df.reset_index(drop=True)
ftr_df = ftr_df.reset_index()
ftr_df
def idx(row):
    return (row['index'], row['exec_group'], "{:02d}".format(row['scenario_id']))

ftr_df['idx'] = ftr_df.apply(idx, axis=1)
ftr_df.drop(columns=['index', 'last_battery_levels', 'ttc', 'last_battery_levels'])
ftr_df.to_csv('failures.csv')

failures = ftr_df['idx'].unique()
failures
```

```python
#prep
tar_df = tar_df.drop(columns=['exec_code', 'parameters'])
def rel_delay(row):
    return row['expent_time'] / row['planned_time']

def abs_delay(row):
    return row['expent_time'] - row['planned_time']

tar_df['rel_delay'] = tar_df.apply(rel_delay, axis=1)
tar_df['abs_delay'] = tar_df.apply(abs_delay, axis=1)
tar_df = tar_df.sort_values(by="start_time").reset_index(drop=True)
```
